```python
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42
)

print("Train:", train_df.shape)
print("Val:", val_df.shape)
```

```python
y_train = train_df["Anxiety Level (1-10)"]
y_val   = val_df["Anxiety Level (1-10)"]
```

```python
X_train_df = train_df.drop(columns=["Anxiety Level (1-10)"])
X_val_df   = val_df.drop(columns=["Anxiety Level (1-10)"])
```

```python
from preprocessing.pipeline import TherapeuticPreprocessor

column_map = {
    "heart_rate": ["Heart Rate (bpm)", "HR"],
    "stress": ["Stress Level", "stress_score"],
    "sleep_hours": ["Sleep Hours"],
    "age": ["Age"],
    "gender": ["Gender"]
}

preprocessor = TherapeuticPreprocessor(column_map)
```

```python
preprocessor.fit(X_train_df)
X_train = preprocessor.transform(X_train_df)
X_val   = preprocessor.transform(X_val_df)

print(X_train.shape)
print(X_val.shape)


print("NaNs in X_train:", X_train.isna().sum())
print("NaNs in X_val:", X_val.isna().sum())
```

```python
X_train_df.columns
```

```python
X_val_df.columns
```

```python
from sklearn.linear_model import Ridge

model = Ridge(alpha=1.0)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

val_preds = model.predict(X_val)
mse = mean_squared_error(y_val, val_preds)

print("Validation MSE:", mse)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_val, val_preds, alpha=0.5)
plt.xlabel("True Anxiety")
plt.ylabel("Predicted Anxiety")
plt.title("Validation Predictions")
plt.show()
```

```python
# test inference shape
sample = X_val.iloc[:1]
pred = model.predict(sample)
print(pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

rf = RandomForestRegressor(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

rf_preds = rf.predict(X_val)
```

```python
mse = mean_squared_error(y_val, rf_preds)
mae = mean_absolute_error(y_val, rf_preds)
r2  = r2_score(y_val, rf_preds)

print("Random Forest Performance")
print("MSE:", mse)
print("MAE:", mae)
print("R² :", r2)
```
